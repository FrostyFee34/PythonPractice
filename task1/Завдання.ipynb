```python
import matplotlib.pyplot as plt
import cv2

img = plt.imread("lines2.jpg")

h, w = img.shape[0:2]
h_new = 400
ratio = w/h
w_new = int(h_new * ratio)
resized = cv2.resize(img, (w_new, h_new))

cv2.putText(resized, 'Valihura', (50,50), cv2.FONT_HERSHEY_DUPLEX, 1, (255, 255, 255), 1)

plt.imshow(resized), plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
import imutils 

img = plt.imread("lines2.jpg")
cv2.putText(img, 'Valihura', (50,50), cv2.FONT_HERSHEY_DUPLEX, 1, (255, 255, 255), 1)
rotated = imutils.rotate(img, -45)


plt.imshow(rotated), plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = plt.imread("lines2.jpg")

blurred = cv2.GaussianBlur(img, (11, 11), 0)
cv2.putText(blurred, 'Valihura', (50,50), cv2.FONT_HERSHEY_DUPLEX, 1, (255, 255, 255), 1)

plt.imshow(blurred), plt.axis('off')
plt.show()
```

```python
import cv2
import imutils 
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((512, 512, 3), np.uint8)

points1 = np.array([[127, 80], [127, 390], [384, 390], [384, 80], [256, 340]])
cv2.polylines(img, np.int32([points1]), 1, (255, 255, 255))

points2 = np.array([[255, 263], [198, 390], [255, 475], [314, 390]])
cv2.polylines(img, np.int32([points2]), 1, (255, 255, 255))

cv2.line(img, (255, 263), (255, 47), (255, 255, 255), 1)

plt.imshow(img), plt.axis('off')
plt.show()
```

```python
import cv2
import imutils 
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread("lines2.jpg")
img1 = plt.imread("building.jpg")

h_new = img.shape[0]
img1 = imutils.resize(img1, height = h_new)
attached = np.hstack((img, img1))

cv2.putText(attached, 'Valihura', (50,50), cv2.FONT_HERSHEY_DUPLEX, 1, (255, 255, 255), 1)

plt.imshow(attached), plt.axis('off')
plt.show()
```
