```python
import dlib
from skimage import io
from scipy.spatial import distance
import cv2

sp = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
facerec = dlib.face_recognition_model_v1('dlib_face_recognition_resnet_model_v1.dat')
detector = dlib.get_frontal_face_detector()

img = io.imread('passport.jpg')
win1 = dlib.image_window()
win1.clear_overlay()
win1.set_image(img)


dets = detector ( img , 1 )
for k, d in enumerate(dets):
    print (" Detection {}: Left : {} Top: {} Right : {} Bottom : {}".format(k, d.left(), d.top(), d.right(), d.bottom()))
    shape = sp(img, d )
    win1.clear_overlay()
    win1.add_overlay(d)
    win1.add_overlay(shape)
    
official_face_descriptor = facerec.compute_face_descriptor(img, shape)

cap = cv2.VideoCapture(0)
ret, frame = cap.read()
cap.release()

win1.clear_overlay()
win1.set_image(frame)

dets1 = detector (frame , 1 )

for k, d in enumerate(dets1):
    print (" Detection {}: Left : {} Top: {} Right : {} Bottom : {}".format(k, d.left(), d.top(), d.right(), d.bottom()))
    shape1 = sp(frame, d )
    for part in shape1.parts():
        cv2.circle(frame, part, 3, (0, 0, 255), 1)
    win1.clear_overlay()
    win1.add_overlay(d)
    win1.add_overlay(shape1)
    
webcam_face_descriptor = facerec.compute_face_descriptor(frame, shape1)
a = distance.euclidean(official_face_descriptor, webcam_face_descriptor)

if(a < 0.6):
    cv2.putText(frame, 'True', (20,30), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 0), 2)
    cv2.putText(frame, 'True', (20,30), cv2.FONT_HERSHEY_DUPLEX, 1, (255, 255, 255), 1)
    cv2.imwrite('imageChecked.jpg', frame)
```

```python
import dlib
import requests
import cv2
from scipy.spatial import distance
from skimage import io
from bs4 import BeautifulSoup

shape_predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
face_rec_model = dlib.face_recognition_model_v1('dlib_face_recognition_resnet_model_v1.dat')
detector = dlib.get_frontal_face_detector()

def getLinks(url):
    page_data = requests.get(url).text
    images_urls = [
        image.attrs.get('src')
        for image in BeautifulSoup(page_data , 'html.parser').find_all('img')
    ]

    images_urls = [
        image_url
        for image_url in images_urls
        if image_url and len ( image_url )>0
    ]
    return images_urls

def getDescriptor(img):
    dets = detector (img , 1)
    for k, d in enumerate(dets):
        print (" Detection {}: Left : {} Top: {} Right : {} Bottom : {}".format(k, d.left(), d.top(), d.right(), d.bottom()))
        shape = shape_predictor(img, d)
    face_descriptor = face_rec_model.compute_face_descriptor(img, shape)
    return face_descriptor
    
    
def formDictionary(url):
    links = getLinks(url)
    dict = {}
    
    for link in links:
        img = io.imread(link)
        desc = getDescriptor(img)
        surname = link[69:-4]
        dict[surname] = desc
    return dict    

def detectEmployees(img, dictionary):
    dets = detector(img, 1)
    for k, d in enumerate(dets):
        shape = shape_predictor(img, d)
        face_descriptor1 = face_rec_model.compute_face_descriptor(img, shape)
        for name, descriptor in dictionary.items():
            dist = distance.euclidean(face_descriptor1, descriptor)
            if dist < 0.60:
                print ('Face detected, '+ name +', distance - ', dist )
                cv2.rectangle(img, (d.left(), d.top()), (d.right(), d.bottom()), (255, 0, 255), 2)
                cv2.putText(img, name, (d.left(),d.top()), cv2.FONT_HERSHEY_DUPLEX, 0.4, (0, 0, 0), 2)
                cv2.putText(img, name, (d.left(),d.top()), cv2.FONT_HERSHEY_DUPLEX, 0.4, (255, 255, 255), 1)

    im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2.imwrite('someimage.jpg', im_rgb)
img = io.imread('http://management.web.khnu.km.ua/wp-content/uploads/sites/16/2017/12/Compressed-N2mRQOqWJ03DtaN-950x550.jpg')
dict_ = formDictionary('http://management.khnu.km.ua/spivrobitnyky/') ## Get the dictionary
detectEmployees(img, dict_) ## Detect employees from dictionary in the image
```
